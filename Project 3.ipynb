# Project 3
## Frankenstein

```python
import string
import matplotlib.pyplot as plt

# Cleaning
file = open('Frankenstein.txt', encoding='utf8')  # Open File
text = file.read()  # Reads file
words_token = text.split()  # Split text / Token
# Removes punctuations & convert to lowercase
punct_char = set(string.punctuation)  # Punctuation varible
clean_text = ''
for char in text:
    if char not in punct_char:
        clean_text += char.lower()

# Word Occurrences
total_words = len(words_token)  # Total Number of Words
unique = set(words_token)  # Set of Unique Words
total_unique = len(unique)  # Total Number of Unique Words

# Dictionary
word_freq = {}
for word in unique:
    word_freq[word] = words_token.count(word)

# Sorts dictionary
def get_freq(item):
    return item[1]

sort_word_freq = dict(sorted(word_freq.items(), key=get_freq, reverse=True))

# Top 30
top30words = []
top30freq = []
for i, word in enumerate(sort_word_freq.keys()):
    if i == 30:
        break
    top30words.append(word)
    top30freq.append(sort_word_freq[word])
   


# Output
print(f'Total Words: {total_words}')
print(f'Total Unique Words: {total_unique}')
print('n\Word\t\t\tFrequency')
for word, freq in sort_word_freq.items():
    print(f'{word}\t\t\t{freq}')
    break

# Data Visualization
plt.figure(figsize=(12, 6))  # Sets width and height of the figure
plt.bar(top30words, top30freq)  # What will be measured
plt.title('Top 30 Most Used Words in Frankenstein',
          fontsize=16)  # Title of visualizatioin
plt.xlabel('Words', fontsize=12)  # X-axis label will  be the words
# Y-axiz of the label will be the frequency
plt.ylabel('Frequency', fontsize=12)
# Sets font size and rotation of the x-axis label
plt.xticks(rotation=90, fontsize=10)

# Labels the exact number of words that occur
for i in range(len(top30words)):
    plt.text(i, top30freq[i] + 50, str(top30freq[i]),
             ha='center', fontsize=8)

plt.show()  # Plots the graph
```
